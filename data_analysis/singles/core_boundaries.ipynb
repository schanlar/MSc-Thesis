```python
import mesa_reader as mr
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
p = mr.MesaData('../3p1M/profile_after_He_depletion.data')
```

```python
fig, ax1 = plt.subplots(figsize = (13,9))


color = 'tab:red'
ax1.set_xlabel('Mass coordinate [M$_{\odot}$]', fontsize = 13)
ax1.set_ylabel('Z mass fraction', color=color, fontsize = 13)
ax1.plot(p.data('mass'), p.data('z_mass_fraction_metals'), color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.get_yaxis().get_major_formatter().set_useOffset(False)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel(r'$\log(P)$ [Ba]', color=color, fontsize = 13)  # we already handled the x-label with ax1
ax2.plot(p.data('mass'), p.data('logP'), color=color)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylim(0, 30)

plt.fill_betweenx([0,30], 1.37, 1.43, alpha = 0.1, color = 'grey')
#plt.axvline(x = 1.384, color = 'grey', linestyle = '--')

fig.tight_layout()  # otherwise the right y-label is slightly clipped

#plt.savefig('../data_analysis/singles_plots/3p1_core_def.png', bbox_inches = 'tight', dpi = 300)
```
